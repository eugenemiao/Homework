## Latitude - Longitude Graphs

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from config import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "Resources/cities.csv"
```

```python
cities_weather = pd.read_csv(output_data_file)

city_id = cities_weather["City_ID"]
city = cities_weather["City"]
cloudiness = cities_weather["Cloudiness"]
country = cities_weather["Country"]
date = cities_weather["Date"]
humidity = cities_weather["Humidity"]
lat = cities_weather["Lat"]
lng = cities_weather["Lng"]
max_temp = cities_weather["Max Temp"]
wind_speed = cities_weather["Wind Speed"]

cities_weather.head()
```

```python
# City Latitude vs Max Temp

plt.scatter(lat, 
            max_temp,
            linewidths=1, marker="o", 
            alpha=0.8, label="Cities")

# Incorporate the other graph properties
plt.title("City Latitude vs. Max Temperature (%s)" % time.strftime("%x"))
plt.ylabel("Max Temperature (F)")
plt.xlabel("Latitude")
plt.grid(True)

# Save the figure
plt.savefig("Output Data/City Latitude vs Max Temp.png")

# Show plot
plt.show()
```
